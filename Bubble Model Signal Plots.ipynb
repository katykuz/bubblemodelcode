# Plotting Acoustic Signals for Bubble Model #
## Description ##
This code plots the acoustic signals (in terms of pressure) that were simulated in COMSOL for the bubble modeling. The signals considered were a continuous wave, half cycle, 1 cycle,  5 cycles, and 10 cycles. An envelope function was used to define the number of cycles from the continuous wave. 

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

## Frequency Parameters ##

```python
f = 7e5 #Hz
T = 1/f
w = 2*np.pi*f #angular frequency
```

### Initial Pressures ##

```python
################# Intial Pressures (Pascals) ##################
p0= 14591 #Pascals, 14591 is 0.144 atmospheres
p1= 100000 
p2= 200000
p3= 500000 #Pa, 500kPa, 0.5 Mpa
p4= 650000
p5= 5000000
```

### Number of Cycles ##

```python
################# Cycle Definition ##################
N05 =0.5 #number of cycles
N1 =1 
N5 = 5
N10 = 10
m = 1 # slope of pulse envelope, 1 is Gaussian 
```

### Time Range ##

```python
################# Time Range ##################
t0 = 0    # initial time
tf = 0.00005   # final time
dt = 0.000000005  # time steps
t = np.arange(t0, tf, dt)
```

### Continuous Acoustic Signal ##

```python
################# Continuous Acoustic Signal ##################
a = np.sin(2*np.pi*f*t) 
```

### Envelope Functions with Different Cycle ##

```python

E_N05= np.exp(-((w*t)/(N05*np.pi))**(2*m))                 #Half Cycle
E_N1= np.exp(-((w*t)/(N1*np.pi))**(2*m))                   #One Cycle
E_N5= np.exp(-((w*t)/(N5*np.pi))**(2*m))                   #Five Cycles
E_N10= np.exp(-((w*t)/(N10*np.pi))**(2*m))                 #Ten Cycles 
```

### Initial Pressures ##

```python
p_N05_p0 = p0*a*E_N05                                      #Half Cycle
p_N1_p0 = p0*a*E_N1                                        #One Cycle
p_N5_p0 = p0*a*E_N5                                        #Five Cycles
p_N10_p0 = p0*a*E_N10                                      #Ten Cycles
```

### Output Signal - 1 Initial Pressure, Different Cycles ##

```python
p_N05_p0 = p0*a*E_N05                                      #Half Cycle
p_N1_p0 = p0*a*E_N1                                        #One Cycle
p_N5_p0 = p0*a*E_N5                                        #Five Cycles
p_N10_p0 = p0*a*E_N10                                      #Ten Cycles
```

### Output Signal - 1 Cycle, Different Initial Pressures ##

```python
p_N05_p0 = p0*a*E_N05                           #Initial Value of 14591 Pa
p_N05_p1 = p1*a*E_N05                           #Initial Value of 100 kPa
p_N05_p2 = p2*a*E_N05                           #Initial Value of 200 kPa
p_N05_p3 = p3*a*E_N05                           #Initial Value of 500 kPa
p_N05_p4 = p4*a*E_N05                           #Initial Value of 650 kPa
p_N05_p5 = p5*a*E_N05                           #Initial Value of 5 MPa
```

### Plotting Continuous Signal ###

```python
plot1 = plt.figure(1)
plt.plot(t, a)
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Continuous Signal')
```

###  Plotting Different Cycle Signals

```python
################# Plotting Different Cycle Signals #######################
plot2 = plt.figure(2)
plt.plot(t, p_N05_p0,'tab:orange')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Half Cycle Signal, Initial Pressure at 14.5kPa')

plot3 = plt.figure(3)
plt.plot(t, p_N1_p0,'tab:green')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('One Cycle Signal, Initial Pressure at 14.5kPa')

plot4 = plt.figure(4)
plt.plot(t, p_N5_p0,'tab:red')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('5 Cycle Signal, Initial Pressure at 14.5kPa')

plot5 = plt.figure(5)
plt.plot(t, p_N10_p0,'tab:purple')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('10 Cycle Signal, Initial Pressure at 14.5kPa')
```

### Plotting Different Initial Pressure Signals

```python
################# Plotting Different Initial Pressure Signals #######################
plot6 = plt.figure(6)
plt.plot(t, p_N05_p0,'tab:orange')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Initial Pressure at 14.5kPa')

plot7 = plt.figure(7)
plt.plot(t, p_N05_p1,'tab:green')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Initial Pressure at 100kPa')

plot8 = plt.figure(8)
plt.plot(t, p_N05_p2,'tab:red')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Initial Pressure at 200kPa')

plot9 = plt.figure(9)
plt.plot(t, p_N05_p3,'tab:purple')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Initial Pressure at 500kPa')

plot10 = plt.figure(10)
plt.plot(t, p_N05_p4,'tab:blue')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure (Pascals)')
plt.title('Initial Pressure at 650kPa')
```
